```python
import pandas as pd
import glob

def merge_excel_files(folder_path):
    # Use glob to find all Excel files in the specified folder
    file_pattern = f"{folder_path}/*.xlsx"  # Change to .xlsv if needed
    all_files = glob.glob(file_pattern)

    # Initialize an empty list to hold DataFrames
    df_list = []

    # Loop through each file and read it into a DataFrame
    for file in all_files:
        df = pd.read_excel(file)
        df_list.append(df)  # Add the DataFrame to the list

    # Concatenate all DataFrames in the list into a single DataFrame
    df_final = pd.concat(df_list, ignore_index=True)

    # After merging all files into df_final
    output_file_path = 'dataset_transcripts/merged_transcripts.xlsx'  # Change this to your desired output path

    # Save the DataFrame to an Excel file
    df_final.to_excel(output_file_path, index=False)


    return df_final

# Specify the folder containing your Excel files
folder_path = "dataset_transcripts"

# Merge all files and store the result in df_final
df_final = merge_excel_files(folder_path)

# Display the final DataFrame
print(df_final)

```
